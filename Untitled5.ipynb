```python
from collections import deque
import matplotlib.pyplot as plt
import networkx as nx

def bfs_shortest_path(start, goal, graph):
    # Queue to store paths, starting with the initial city
    queue = deque([[start]])
    # Set to keep track of visited cities to avoid cycles
    visited = set([start])

    while queue:
        # Get the first path from the queue
        path = queue.popleft()
        # Get the last city in the path
        city = path[-1]

        # If we reached the goal city, return the path
        if city == goal:
            return path

        # Explore connected cities
        for neighbor in graph[city]:
            if neighbor not in visited:
                visited.add(neighbor)  # Mark as visited
                new_path = path + [neighbor]  # Create a new path
                queue.append(new_path)  # Add to the queue

    # If no path is found, return None
    return None

# Graph representation of the cities and connections
graph = {
    "Islamabad": ["Rawalpindi", "Lahore", "Peshawar"],
    "Rawalpindi": ["Islamabad", "Peshawar", "Quetta"],
    "Peshawar": ["Islamabad", "Rawalpindi", "Quetta"],
    "Lahore": ["Islamabad", "Multan", "Quetta"],
    "Multan": ["Lahore", "Karachi", "Quetta"],
    "Quetta": ["Rawalpindi", "Peshawar", "Multan", "Karachi"],
    "Karachi": ["Multan", "Quetta"]
}

# Find the shortest path from Islamabad to Karachi
shortest_path = bfs_shortest_path("Islamabad", "Karachi", graph)
print("Shortest path from Islamabad to Karachi:", shortest_path)

# Create a NetworkX graph from the city connections
G = nx.Graph()
for city, neighbors in graph.items():
    for neighbor in neighbors:
        G.add_edge(city, neighbor)

# Draw the graph using NetworkX and Matplotlib
pos = nx.spring_layout(G)  # Position the nodes using the spring layout
plt.figure(figsize=(10, 8))

# Draw all nodes and edges
nx.draw_networkx_nodes(G, pos, node_size=5000, node_color="orange")
nx.draw_networkx_edges(G, pos, width=2, edge_color="gray")
nx.draw_networkx_labels(G, pos, font_size=12, font_weight="bold")
nx.draw_networkx_nodes(G, pos, nodelist=shortest_path, node_color='lightgreen', node_size=5000)


# Highlight the shortest path in red (if found)
if shortest_path:
    path_edges = list(zip(shortest_path, shortest_path[1:]))  # Create pairs of consecutive cities
    nx.draw_networkx_edges(G, pos, edgelist=path_edges, width=4, edge_color="red")

# Add title and display the graph
plt.title("City Connections and Shortest Path", fontsize=15)
plt.axis("off")  # Hide axes
plt.show()
```
